```python
X = df.iloc[:, :1]
y = df.iloc[:, 1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import  LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, y_pred)
```

```python
class GD:
  def __init__(self, lr=0.01, epochs=1000):
    self.lr = lr
    self.epochs = epochs
    self.coef_ = 100
    self.intercept_ = 100

  def fit(self, X_train, y_train):
    for i in range(self.epochs):
      self.slope = 0
      for j in range(len(X_train)):
        self.slope = self.slope + (y_train[j] - self.coef_ * X_train[j] - self.intercept_)
      self.slope = -2 * self.slope
      self.intercept_ = self.intercept_ - (self.lr * self.slope)

      self.slope = 0
      for j in range(len(X_train)):
        self.slope = self.slope + (y_train[j] - self.coef_ * X_train[j] - self.intercept_) * X_train[j]
      self.slope = -2 * self.slope
      self.coef_ = self.coef_ - (self.lr * self.slope)

      plt.plot(X_train, self.coef_ * X_train + self.intercept_)
      print(self.coef_, self.intercept_)
    plt.scatter(X_train, y_train)
```
